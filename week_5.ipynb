```python
import socket
import datetime  # Importing the datetime library

# Create a TCP socket
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(('localhost', 65432))  # Bind the server to localhost and port 
server_socket.listen(1)  

print("TCP Server is listening...")

while True:
    # Accept incoming client connections
    client_socket, client_address = server_socket.accept()
    print(f"Connected to {client_address}")
    
    # Track the start time
    start_time = datetime.datetime.now()
    
    # Receive data from the client
    data = client_socket.recv(1024)
    print(f"Received: {data.decode()}")
    
    # Echo back the data with "ACK:" prefix
    client_socket.sendall(b"ACK: " + data)
    
    # Track the end time
    end_time = datetime.datetime.now()
    
    # Calculate the time taken 
    time_taken = end_time - start_time
    print(f"Time taken : {time_taken}")
    
    # Close the client socket
    client_socket.close()
```

```python
import socket
import datetime  # Importing the datetime library

# Create a TCP socket for server
tcp_server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
tcp_server_socket.bind(('localhost', 65432))
tcp_server_socket.listen(1)

# Create a UDP socket for server
udp_server_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
udp_server_socket.bind(('localhost', 65433))  # Use a different port for UDP

print("TCP Server is listening...")
print("UDP Server is listening...")

while True:
    # TCP Server handling
    tcp_client_socket, tcp_client_address = tcp_server_socket.accept()
    print(f"TCP Connected to {tcp_client_address}")

    # Track the start time for TCP
    tcp_start_time = datetime.datetime.now()

    # Receive data from TCP client
    tcp_data = tcp_client_socket.recv(1024)
    print(f"TCP Received: {tcp_data.decode()}")

    # Echo back the data for TCP
    tcp_client_socket.sendall(b"ACK: " + tcp_data)

    # Track the end time for TCP
    tcp_end_time = datetime.datetime.now()
    tcp_time_taken = tcp_end_time - tcp_start_time
    print(f"TCP Time taken: {tcp_time_taken}")

    tcp_client_socket.close()

    # UDP Server handling
    udp_data, udp_client_address = udp_server_socket.recvfrom(1024)
    print(f"UDP Received from {udp_client_address}: {udp_data.decode()}")

    # Track the start time for UDP
    udp_start_time = datetime.datetime.now()

    # Send acknowledgment for UDP (using sendto())
    udp_server_socket.sendto(b"ACK: " + udp_data, udp_client_address)

    # Track the end time for UDP
    udp_end_time = datetime.datetime.now()
    udp_time_taken = udp_end_time - udp_start_time
    print(f"UDP Time taken: {udp_time_taken}")
```

```python
import socket
import datetime  # Importing the datetime library

# Create a TCP socket for client
tcp_client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
tcp_client_socket.connect(('localhost', 65432))

# Create a UDP socket for client
udp_client_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# Track the start time for TCP
tcp_start_time = datetime.datetime.now()

# Input message from the user for TCP
message = input("Enter message for TCP and UDP: ")

# Send the message to the TCP server
tcp_client_socket.sendall(message.encode())

# Receive the server's response (TCP)
tcp_response = tcp_client_socket.recv(1024)

# Track the end time for TCP
tcp_end_time = datetime.datetime.now()
tcp_time_taken = tcp_end_time - tcp_start_time
print(f"TCP Server response: {tcp_response.decode()}")
print(f"TCP Time taken for round trip: {tcp_time_taken}")

# Close TCP socket
tcp_client_socket.close()

# Track the start time for UDP
udp_start_time = datetime.datetime.now()

# Send the message to the UDP server
udp_client_socket.sendto(message.encode(), ('localhost', 65433))

# Receive the server's response (UDP)
udp_response, _ = udp_client_socket.recvfrom(1024)

# Track the end time for UDP
udp_end_time = datetime.datetime.now()
udp_time_taken = udp_end_time - udp_start_time
print(f"UDP Server response: {udp_response.decode()}")
print(f"UDP Time taken for round trip: {udp_time_taken}")

# Close UDP socket
udp_client_socket.close()
```

```python
import socket
import datetime  # Importing the datetime library

# Create a TCP socket
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(('localhost', 65432))  # Bind to localhost and port 65432
server_socket.listen(1)

print("TCP Server is listening...")

while True:
    client_socket, client_address = server_socket.accept()
    print(f"Connected to {client_address}")

    # Track the start time for receiving data
    start_time = datetime.datetime.now()

    # Open the file to save the received data
    with open('received_file.txt', 'wb') as f:
        while True:
            data = client_socket.recv(1024)  # Receive data in chunks of 1024 bytes
            if not data:
                break  # Stop when there's no more data
            f.write(data)  # Write the received data to the file

    # Track the end time for saving data
    end_time = datetime.datetime.now()
    time_taken = end_time - start_time
    print(f"Data saved to file in {time_taken} seconds!")

    # Close the client connection
    client_socket.close()
```

```python
 # server code

import socket
import threading

# List of connected clients
clients = []

# Broadcast message to all clients
def broadcast(message, client_socket):
    for client in clients:
        if client != client_socket:
            try:
                client.send(message)
            except:
                remove_client(client)

# Remove a client from the list
def remove_client(client_socket):
    if client_socket in clients:
        clients.remove(client_socket)

# Handle communication with a client
def handle_client(client_socket):
    while True:
        try:
            message = client_socket.recv(1024)
            if not message:
                break
            print(f"Message received: {message.decode()}")
            broadcast(message, client_socket)
        except:
            break

    remove_client(client_socket)
    client_socket.close()

# Main server function
def start_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind(('localhost', 12345))
    server_socket.listen(5)
    print("Server started. Waiting for clients...")

    while True:
        client_socket, client_address = server_socket.accept()
        print(f"New connection from {client_address}")
        clients.append(client_socket)

        # Start a new thread to handle this client
        client_thread = threading.Thread(target=handle_client, args=(client_socket,))
        client_thread.start()

# Run the server
if __name__ == "__main__":
    start_server()
```

```python
# client code 

import socket
import threading

# Function to receive messages from the server
def receive_messages(client_socket):
    while True:
        try:
            message = client_socket.recv(1024)
            print(f"Message from server: {message.decode()}")
        except:
            print("Connection lost.")
            break

# Function to send messages to the server
def send_messages(client_socket):
    while True:
        message = input("Enter message: ")
        client_socket.send(message.encode())

# Main function
def start_client():
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_socket.connect(('localhost', 12345))

    # Start receiving and sending messages in separate threads
    receive_thread = threading.Thread(target=receive_messages, args=(client_socket,))
    receive_thread.start()

    send_thread = threading.Thread(target=send_messages, args=(client_socket,))
    send_thread.start()

# Run the client
if __name__ == "__main__":
    start_client()
```
